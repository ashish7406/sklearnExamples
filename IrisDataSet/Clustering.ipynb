## K_means 

* It is an Unsupervised algorithm as it has no prediction variables

* It will just finds the pattern in the data

* How it works: 1) It will assign each data point randomly to some clusters 2) Then it will move centroid of each cluster

* We continue this untill : `within cluster variation` cannot be reduced any further

* `within cluster variation is calculated as the sum of euclidean distance between the data points and their respective cluster centroids`

### Import the Libraries:

```python
import warnings
warnings.filterwarnings("ignore")


import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

###### We are going to create Artificial data using scikit-learn :

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
# data: have a look at the output : we will be having two arrays one for features and one for labels and hence it's a tuple

# data[0] >>>>>>>>> gives the features

# data[1] >>>>>>>>> gives the labels
```

```python
data[1].shape
```

```python
data[0].shape
```

### Blobs in our case are nothing but clusters 

#### Visualize :

```python
plt.scatter(data[0][:,0] , data[0][:,1])
```

#### The above plot looks like we have only two groups. But these data points can form 4 clusters: 

```python
plt.scatter(data[0][:,0] , data[0][:,1] , c=data[1] , cmap='rainbow')
```

### Creating Clusters :

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_    
```

###### Since K_means is a Unsupervised algorithm we don't have prediction variables. So we have to predict the labels and no need to comapre the predicted labels to the actual labels because we don't have actual labels

###### But  here we have created Artificial data and we already have actual labels. So for this workbook we can actually compare actual and predicted labels

#### Comparing the Results :

```python
fig , (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_ , cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
